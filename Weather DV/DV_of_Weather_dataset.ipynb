```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("/content/weather_dataset_large.csv",encoding="utf-8")

# Display basic statistics
print(df.describe())

# Check for missing values
print("Missing values:\n", df.isnull().sum())

# Convert date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

# Pairplot for multivariate visualization
sns.pairplot(df, hue="Weather Condition")
plt.show()

# Line plot for temperature trends
df.set_index("Date")["Temperature (°C)"].plot(figsize=(12, 5), title="Temperature Trend")
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.show()

# Box plot for temperature distribution per weather condition
plt.figure(figsize=(10, 6))
sns.boxplot(x="Weather Condition", y="Temperature (°C)", data=df)
plt.xticks(rotation=45)
plt.title("Temperature Distribution by Weather Condition")
plt.show()
```
